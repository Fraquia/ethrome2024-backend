# Synthetic Data Generation

Using statistic about fraud's data we can generate new synthetic data according to features original distribution. This is made in order to make the proportion of Fraud and Not-Fraud row up to 40% and 60% and then train a model more capable or recognize the difference between the two classes

```python
import numpy as np

def generate_synthetic_data(mean, std_dev, min_val, max_val, num_samples):
    # Generate synthetic data using a normal distribution
    data = np.random.normal(loc=mean, scale=std_dev, size=num_samples)

    # Clip the data to ensure it falls within the specified min and max
    data = np.clip(data, min_val, max_val)
    
    return data
```

```python
synthetic_data_dict = {}
num_samples = 1009      # new Fraud Data

for index, row in not_fraud_stats.iterrows():
    if index=='Fraud':
        synthetic_data_dict[index] = np.asarray([1 for x in range(1009)])
    else:
        count, mean, std, _min, _25, _50, _75, _max = row.values
        new_data = generate_synthetic_data(mean, std, _min, _max, num_samples)
        synthetic_data_dict[index] = new_data
```

```python
fraud_synth_data_df = pd.DataFrame.from_dict(synthetic_data_dict)
final_dataset = pd.concat([not_fraud_df.drop(columns="Title"), fraud_synth_data_df])
final_dataset.drop(columns=["Temporal details", "Spatial details", "Social details", 
                            "Perceptual details",'Negation',"Causation"], inplace=True)
final_dataset.head()
```

```python
df_normalized = (final_dataset.drop(columns=['Fraud']) - final_dataset.drop(columns=['Fraud']).min()) / (final_dataset.drop(columns=['Fraud']).max() - final_dataset.drop(columns=['Fraud']).min())
df_normalized['Fraud']=final_dataset['Fraud']
df_normalized
```

```python
#fraud_synth_data_df.drop(columns=["Temporal details", "Spatial details", "Social details", "Perceptual details"], inplace=True)
fraud_synth_data_df.drop(columns=['Negation',"Causation"], inplace=True)
fraud_synth_data_df.info()
```

## Training 

```python
# cretaing x and y dataset
y = df_normalized.Fraud.values
x = df_normalized.drop(columns=['Fraud'])

# splitting x and y in train and test
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

# Create the DMatrix (XGBoost specific format)
dtrain = xgb.DMatrix(x_train, label=y_train)
dtest = xgb.DMatrix(x_test, label=y_test)
```

```python
print(type(x_train))
print(x_train.shape)
```

```python
params = {
    'objective': 'binary:logistic',
    'max_depth': 4,
    'learning_rate': 0.01,
    'n_estimators': 150,
    'eval_metric': 'logloss'
}

# Train the model
bst = xgb.train(params, dtrain, num_boost_round=100)
```

```python
# Make predictions
preds = bst.predict(dtest)
predictions = [round(value) for value in preds]
```

```python
accuracy = accuracy_score(y_test, predictions)
print(f"Accuracy: {accuracy * 100:.2f}%")
```

# Saving Model
